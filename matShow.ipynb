### plotting patches with different intensities ###

Note the use of parameters *vmin* and *vmax* when using 
method imshow()

patches of images of a constant grey scale are created on a grid
of nrows , ncols

patches are numbered by id := {0, ..., nr. of patches}

```python
nrows=5
ncols=2
nfig = nrows * ncols
incr = 255 / nfig
id = 0

# id runs from 0 to (nfig-1) 
# and serves to create grey values {0, incr, 2*incr, ..., (nfig-1)*incr}
# resulting in patches with decreasing darkness

fig, axes = plt.subplots(nrows, ncols, figsize=(7, 10), squeeze=True)
for nr in range(nrows):
    for nc in range(ncols):
        greyValue = id*incr
        mm = np.full( (20, 40), greyValue, dtype=np.uint8)
        axes[nr, nc].imshow(mm, cmap='gray', vmin=0, vmax=255)
        axes[nr, nc].axis('off');
        id += 1
```
